**5.Check and handle outliers in at least 3 columns in the dataset**

```python
selected_columns=['Age','Fare','SibSp']
selected_columns
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=data1[selected_columns])
plt.title('Boxplot for outliers')
plt.show()
```

```python
#AGE
Q1=np.percentile(data1['Age'],25,method='midpoint')
Q2=np.percentile(data1['Age'],50,method='midpoint')
Q3=np.percentile(data1['Age'],75,method='midpoint')
print("Q1 is : ",Q1)
print("Q2 is : ",Q2)
print("Q3 is : ",Q3)
```

```python
IQR=Q3-Q1
print(IQR)
low_lim=Q1-1.5*IQR
up_lim=Q3+1.5*IQR
print('lower value:',low_lim)
print("upper value: ",up_lim)    
```

```python
#Identify outliers using IQR

outliers_rows=((data1['Age']<(low_lim))|(data1['Age']<(up_lim)))
print("Rows with outliers:")
print(outliers_rows)
```

```python
outliers=[]
for x in data1['Age']:
  if((x<low_lim)or (x>up_lim)):
    outliers.append(x)
                    
print(outliers)

plt.boxplot(data1['Age'])
plt.show();
```

```python
#FARE
Q1=np.percentile(data1['Fare'],25,method='midpoint')
Q2=np.percentile(data1['Fare'],50,method='midpoint')
Q3=np.percentile(data1['Fare'],75,method='midpoint')
print("Q1 is : ",Q1)
print("Q2 is : ",Q2)
print("Q3 is : ",Q3)

IQR=Q3-Q1
print(IQR)
low_lim=Q1-1.5*IQR
up_lim=Q3+1.5*IQR
print('lower value:',low_lim)
print("upper value: ",up_lim)

#Identify outliers using IQR

outliers_rows=((data1['Fare']<(low_lim))|(data1['Fare']>(up_lim)))
print("Rows with outliers:")
print(outliers_rows)
```

```python
data1['Fare']=data1['Fare'].clip(lower=low_lim,upper=up_lim)

outliers=[]
for x in data1['Fare']:
  if((x<low_lim)or (x>up_lim)):
    outliers.append(x)
                    
print(outliers)

plt.boxplot(data1['Fare'])
plt.show();
```

```python
#SibSp
Q1=np.percentile(data1['SibSp'],25,method='midpoint')
Q2=np.percentile(data1['SibSp'],50,method='midpoint')
Q3=np.percentile(data1['SibSp'],75,method='midpoint')
print("Q1 is : ",Q1)
print("Q2 is : ",Q2)
print("Q3 is : ",Q3)

IQR=Q3-Q1
print(IQR)
low_lim=Q1-1.5*IQR
up_lim=Q3+1.5*IQR
print('lower value:',low_lim)
print("upper value: ",up_lim)

#Identify outliers using IQR

outliers_rows=((data1['SibSp']<(low_lim))|(data1["SibSp"]>(up_lim)))
print("Rows with outliers:")
print(outliers_rows)

```

```python
data1['SibSp']=data1['SibSp'].clip(lower=low_lim,upper=up_lim)

outliers=[]
for x in data1['SibSp']:
  if((x<low_lim)or (x>up_lim)):
    outliers.append(x)
                    
print(outliers)

plt.boxplot(data1['SibSp'])
plt.show();
```

```python
dt=dt.drop(['Name','Ticket'],axis=1)
dt
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
dt['Embarked']=le.fit_transform(dt['Embarked'])
dt
```

```python
dt=pd.get_dummies(dt).astype(int)  #onehot encoding in Gender
```

```python
dt
```

```python
sns.heatmap(dt.corr(),annot=True);
```

```python
X=dt.drop('Survived',axis=1)   #indep variables
y=dt['Survived']    
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.25)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k=[]
neighbors=np.arange(3,30)
neighbors
```

```python
for k in neighbors:
    classifier=KNeighborsClassifier(n_neighbors=k,metric='minkowski',p=2)
    classifier.fit(X_train,y_train)
    knn_pred=classifier.predict(X_test)
    accuracy =accuracy_score(y_test,knn_pred)
    metric_k.append(accuracy)
```

```python
plt.plot(neighbors,metric_k,'o-')
plt.xlabel("neighbors")
plt.ylabel("Metric K")
```

```python
classifier=KNeighborsClassifier(n_neighbors=9,metric='minkowski',p=2)
classifier.fit(X_train,y_train)
knn_pred=classifier.predict(X_test)
```

```python
confusion_matrix(y_test,knn_pred)
```

```python
print("Accuracy: ",accuracy_score(y_test,knn_pred))
# print("Precision: ",precision_score(y_test,knn_pred))
# print("Recall: ",recall_score(y_test,knn_pred))
# print("F1: ",f1_score(y_test,knn_pred))
```

```python
from sklearn.model_selection import KFold
kfold_validator=KFold(10)   #mention how many folds to be created (total data will be divided into k folds)  k values can be 3,5,or 10 depending on the size of data
```

```python
for train_index,test_index in kfold_validator.split(X,y):
    print("Training index: ",train_index)
    print("Testing index: ",test_index)
```

```python
from sklearn.model_selection import cross_val_score
cv_result=cross_val_score(sv_clf,X,y,cv=kfold_validator)
```

```python
from sklearn.model_selection import StratifiedKFold
strat_validator=StratifiedKFold(n_splits=10)     #split-> no of folds data is divided into
```

```python
for train_index,test_index in strat_validator.split(X,y):
    print("Training index: ",train_index)
    print("Testing index: ",test_index)     #values are taken in test and train accoring to the weightage .
```

```python
skr_cv_result=cross_val_score(sv_clf,X,y,cv=strat_validator)
```

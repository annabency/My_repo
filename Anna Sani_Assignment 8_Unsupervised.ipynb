```python
from sklearn.cluster import KMeans
```

```python
X=data.iloc[:,[3,4]].values 
```

```python
wcss=[] #(with in clusters sum of squares)
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
```

```python
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=42)
```

```python
y_kmeans=kmeans.fit_predict(X)
```

```python
X
```

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=100,c='red',label="Cluster 01")
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=100,c='blue',label="Cluster 02")
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=100,c='green',label="Cluster 03")
plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],s=100,c='yellow',label="Cluster 04")
plt.title('cluster of wine')
plt.xlabel('Ash_Alcanity')
plt.ylabel('Magnesium')
plt.legend()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score_avg=silhouette_score(X,y_kmeans)
silhouette_score_avg
```

**Agglomerative Clustering**

```python
import scipy.cluster.hierarchy as sch
dend=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Wine type')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=3,metric='euclidean',linkage='ward')
y_hc=hc.fit_predict(X)
```

```python
y_hc
```

```python
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='red',label='Cluster1')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='green',label='Cluster2')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='yellow',label='Cluster3')

plt.xlabel('Ash_Alcanity')
plt.ylabel('Magnesium')
plt.legend()
plt.show()
```

```python
X
```

```python
sil=silhouette_score(X,y_hc)
sil
```
